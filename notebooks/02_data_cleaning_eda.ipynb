Clean + EDA.

```python
import pandas as pd

df = pd.read_csv('../data/raw/real_estate_raw_data.csv')
df['date']=pd.to_datetime(df['date'])
df.drop_duplicates(inplace=True)
df['price_per_sqft']=df['price']/df['size_sqft']
df['annual_rent']=df['rent']*12
df.to_csv('../data/processed/real_estate_processed_data.csv', index=False)
print('Processed & saved.')
```
